```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df = df.drop(['Id'],1)
```

# Coefficient Correlation

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

### Positive Correlation

```python
plt.figure(figsize = (10,7))
sns.regplot(x='PetalWidthCm',y='PetalLengthCm',data = df)
plt.show()
```

### Negative Correlation

```python
plt.figure(figsize = (10,7))
sns.regplot(x='SepalWidthCm',y='PetalLengthCm',data = df)
sns.scatterplot(x='SepalWidthCm',y='PetalLengthCm',hue='Species',data = df)
plt.show()
```

### Weak Correlation

```python
plt.figure(figsize = (10,7))
sns.regplot(x='SepalWidthCm',y='SepalLengthCm',data = df)
plt.show()
```

# Mengetahui Sebaran dari SepalLengthCm

### Histogram

```python
plt.figure(figsize = (10,7))
plt.hist(df['SepalLengthCm'])
plt.show()
```

# Mengubah Sebaran tidak Normal menjadi Sebaran Normal

## Transformasi Logaritma Natural

```python
x = np.log(df['SepalLengthCm'])

plt.figure(figsize = (10,7))
plt.hist(x)
plt.show()
```

### Mengapa ada Tiga Puncak?

```python
setosa = df[df['Species'] == 'Iris-setosa']
virginica = df[df['Species'] == 'Iris-virginica']
versicolor = df[df['Species'] == 'Iris-versicolor']
```

```python
plt.figure(figsize = (10,7))
plt.hist(setosa['SepalLengthCm'],label='setosa')
plt.hist(virginica['SepalLengthCm'],label='virginica')
plt.hist(versicolor['SepalLengthCm'],label='versicolor')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,7))
sns.distplot(setosa['SepalLengthCm'],label='setosa')
sns.distplot(virginica['SepalLengthCm'],label='virginica')
sns.distplot(versicolor['SepalLengthCm'],label='versicolor')
plt.legend()
plt.show()
```
