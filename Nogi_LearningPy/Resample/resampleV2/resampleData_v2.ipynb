```python
import pandas as pd
```

```python
# Sheet alex
sheetName=[
    '20220221_MRT',
    '20220222_MRT',
    '20220223_MRT',
    '20220224_MRT',
    '20220228_MRT',
    '20220301_MRT',
    '20220302_MRT',
    '20220303_MRT',
]
# sheet clayton
# sheetName=[
#     '20220223_MRT',
#     '20220224_MRT',
#     '20220228_MRT',
#     '20220301_MRT',
#     '20220302_MRT',
#     '20220303_MRT',
# ]

listData = []

for name in sheetName:
    listData.append(pd.read_excel("ThermalEquipmentDataNA_Alex.xlsx", sheet_name=name))

len(listData)
```

```python
for index, data in enumerate(listData):
    data['time'] = data['Date'].astype(str) + ' ' +data['Time'].astype(str)
    data['time']=data['time'].astype("datetime64[ns]")
    data.set_index("time", inplace=True)
    listData[index]=data.resample('20min', label='right', closed='right').mean().dropna()
```

```python
result=pd.concat(listData)
result
```
