```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("DataforSEMBayesian.xlsx")
dataAdrian = pd.read_excel("Adrianresample.xlsx")
dataAlex = pd.read_excel("Alexresample.xlsx")
dataClayton = pd.read_excel("Claytonresample.xlsx")
```

```python
category = {
    "Acoustic5Scale":{'Not at all':0, 'Slightly':1, 'Moderately':2, 'Very':3, 'Extremely':4},
    "AcousticConsentration":{'Not at all':0, 'Slightly':1, 'Moderately':2, 'Very':3, 'Extremely':4},
    "ThermalSensation":{'Cold':1, 'Cool':2, 'Slightly Cool':3, 'Neutral':4, 'Slightly Warm':5, 'Warm':6, 'Hot':7},
    "SubjectiveAirflow":{'Much Too Still':1, 'Too Still':2, 'Slightly Still':3, 'Just right':4, 'Slightly Breezy':5, 'Too Breezy':6, 'Much Too Breezy':7},
    "SubjectiveHumidity":{'Much Too Dry':1, 'Too Dry':2, 'Slightly Dry':3, 'Just Right':4, 'Slightly Humid':5, 'Too Humid':6, 'Much Too Humid':7},
    "ThermalAcceptability":{'Clearly unacceptable':1, 'Just unacceptable':2, 'Just acceptable':3, 'Clearly acceptable':4}
}
```

```python
data.replace(category, inplace=True)
```

```python
adrianID = [
    'id02',
    'id05',
    'id07',
    'Id07',
    'id10',
    'id13',
    'id16',
    'id19',
    'id22'
]
alexID = [
    'id03',
    'id06',
    'id09',
    'id9',
    'id12',
    'id15',
    'id18',
    'id21',
    'id24'
]
claytonID = [
    'id08',
    'id11',
    'id14',
    'id17',
    'id17 ',
    'Id17',
    'Id17 ',
    'Id 17',
    'id20',
    'id23'
]

V1 = []
RH = []
DB_T = []
Globe_T = []

for index, row in data.iterrows():
    if row['ID'] in adrianID:
        # try:
        dataFiltered = dataAdrian[row['EndTime'] < dataAdrian['time']]        
        V1.append(dataFiltered['V1'].values[:1][0])
        RH.append(dataFiltered['RH'].values[:1][0])
        DB_T.append(dataFiltered['DB T'].values[:1][0])
        Globe_T.append(dataFiltered['Globe T'].values[:1][0])
    elif row['ID'] in alexID:
        dataFiltered = dataAlex[row['EndTime'] < dataAlex['time']]        
        V1.append(dataFiltered['V1'].values[:1][0])
        RH.append(dataFiltered['RH'].values[:1][0])
        DB_T.append(dataFiltered['DB T'].values[:1][0])
        Globe_T.append(dataFiltered['Globe T'].values[:1][0])
    elif row['ID'] in claytonID:
        dataFiltered = dataClayton[row['EndTime'] < dataClayton['time']]        
        V1.append(dataFiltered['V1'].values[:1][0])
        RH.append(dataFiltered['RH'].values[:1][0])
        DB_T.append(dataFiltered['DB T'].values[:1][0])
        Globe_T.append(dataFiltered['Globe T'].values[:1][0])
    else:
        V1.append(np.nan)
        RH.append(np.nan)
        DB_T.append(np.nan)
        Globe_T.append(np.nan)
        
    
data['V1'] = V1
data['RH'] = RH
data['DB T'] = DB_T
data['Globe T'] = Globe_T
```
