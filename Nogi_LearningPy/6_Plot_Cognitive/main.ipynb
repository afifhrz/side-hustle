# Wilcoxon

## Traffic Noise 45 and NV Fan 3 Included

```python
data_thermal = sorted(list(data_combined.Thermal.unique()))
sb_y = ["Rotation_max", "SpatialSpan_max", "DoubleTrouble_max", "FeatureMatch_max"]
data_final_plot4b = data_combined[['Traffic noise levels', 'Thermal']+sb_y]
data_final_plot4b['Thermal'] = pd.Categorical(data_final_plot4b['Thermal'],data_thermal)
data_final_plot4b.sort_values('Thermal', inplace=True)
data_final_plot4b.reset_index(inplace=True)
data_final_plot4b
```

### on a same traffic noise levels different thermal

```python
# Calculate Wilcoxon Significance Test
from itertools import combinations

result = {}

data_traffic_noise = list(data_final_plot4b['Traffic noise levels'].unique())
for traffic_noise in data_traffic_noise:
    if traffic_noise==45:
        continue
    temp_data = data_final_plot4b[data_final_plot4b['Traffic noise levels'] == traffic_noise]
    data_thermal_loop = list(temp_data['Thermal'].unique())
    thermal_combin = list(combinations(data_thermal_loop, 2))
    result[traffic_noise] = {}
    for thermal_x, thermal_y in thermal_combin:
        data_x = temp_data[temp_data['Thermal'] == thermal_x]
        data_y = temp_data[temp_data['Thermal'] == thermal_y]
        result[traffic_noise][thermal_x+"_"+thermal_y] = {}
        for item_ in sb_y:
            final_data_x = data_x[item_].values
            final_data_y = data_y[item_].values
            res = scp.stats.wilcoxon(final_data_x, final_data_y, alternative="two-sided", mode="approx")
            result[traffic_noise][thermal_x+"_"+thermal_y][item_] = res.pvalue
```

```python
import pprint
pprint.pprint(result)
```

### on a same thermal different noise levels

```python
# Calculate Wilcoxon Significance Test
from itertools import combinations

result_2 = {}

data_thermal_loop = list(data_final_plot4b['Thermal'].unique())
for thermal_ in data_thermal_loop:
    temp_data = data_final_plot4b[data_final_plot4b['Thermal'] == thermal_]
    data_traffic_loop = list(temp_data['Traffic noise levels'].unique())
    traffic_combin = list(combinations(data_traffic_loop, 2))
    result_2[thermal_] = {}
    for traffic_x, traffic_y in traffic_combin:
        data_x = temp_data[temp_data['Traffic noise levels'] == traffic_x]
        data_y = temp_data[temp_data['Traffic noise levels'] == traffic_y]
        result_2[thermal_][str(traffic_x)+"_"+str(traffic_y)] = {}
        for item_ in sb_y:
            final_data_x = data_x[item_].values
            final_data_y = data_y[item_].values
            res = scp.stats.wilcoxon(final_data_x, final_data_y, alternative="two-sided", mode="approx")
            result_2[thermal_][str(traffic_x)+"_"+str(traffic_y)][item_] = res.pvalue
```

## NV Fan 5 Vs AC

```python
data_thermal = sorted(list(data_combined.Thermal.unique()))
sb_y = ["Rotation_max", "SpatialSpan_max", "DoubleTrouble_max", "FeatureMatch_max"]
data_final_plot4b = data_combined[['Traffic noise levels', 'Thermal']+sb_y]
data_final_plot4b['Thermal'] = pd.Categorical(data_final_plot4b['Thermal'],data_thermal)
data_final_plot4b.sort_values('Thermal', inplace=True)
data_final_plot4b.reset_index(inplace=True)
data_final_plot4b
```

```python
# Calculate Wilcoxon Significance Test

result_3 = {}
result_3['55+70'] = {}
col_x = 'NV fan 5'
col_y = 'AC'

data_x = data_final_plot4b[(data_final_plot4b['Thermal'] == col_x) & ~(data_final_plot4b['Traffic noise levels'] == 45)]
data_y = data_final_plot4b[(data_final_plot4b['Thermal'] == col_y) & ~(data_final_plot4b['Traffic noise levels'] == 45)]
result_3['55+70'][col_x+"_"+col_y] = {}
for item_ in sb_y:
    final_data_x = data_x[item_].values
    final_data_y = data_y[item_].values
    res = scp.stats.wilcoxon(final_data_x, final_data_y, alternative="two-sided")
    result_3['55+70'][col_x+"_"+col_y][item_] = res.pvalue
```

```python
pprint.pprint(result_3)
```

```python
import matplotlib.pyplot as plt
# Make figures A6 in size
A = 6
plt.rc('figure', figsize=[46.82 * .5**(.5 * A), 33.11 * .5**(.5 * A)])
# Change figure quality
plt.rc('figure', dpi=141)
# Use LaTeX for graphs' text
plt.rc('text', usetex=False)
# Use the serif font
plt.rc('font', family='serif')

for item_ in sb_y:
    # Create a set of axes
    ax = plt.axes()
    # Create a boxplot on the axes
    bp = ax.boxplot([data_x[item_], data_y[item_]], widths=0.6, patch_artist=True)
    # Graph title
    ax.set_title("Thermal type VS "+item_, fontsize=16)
    # Label y-axis
    ax.set_ylabel(item_)
    # Label x-axis ticks
    xticklabels = [col_x, col_y]
    ax.set_xticklabels(xticklabels)
    # Hide x-axis major ticks
    ax.tick_params(axis='x', which='major', length=0)
    # Show x-axis minor ticks
    xticks = [0.5] + [x + 0.5 for x in ax.get_xticks()]
    ax.set_xticks(xticks, minor=True)
    # Clean up the appearance
    ax.tick_params(axis='x', which='minor', length=3, width=1)

    # Change the colour of the boxes to Seaborn's 'pastel' palette
    colors = [(1,0,0),(0,0,1)]
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)

    # Colour of the median lines
    plt.setp(bp['medians'], color='k')

    # Get the y-axis limits
    bottom, top = ax.get_ylim()
    y_range = top - bottom

    # Significance bars

    # What level is this bar among the bars above the plot?
    level = 1
    
    # Plot the bar
    bar_height = (y_range * 0.07 * level) + top
    bar_tips = bar_height - (y_range * 0.02)
    plt.plot(
        [1, 1, 2, 2],
        [bar_tips, bar_height, bar_height, bar_tips], lw=1, c='k'
    )

    # extending y axis
    plt.axline(xy1=(1, top + (.1 * top)), xy2=(2, .5 * y_range), color=(0, 0, 0, 0))

    # Significance level
    p = result_3['55+70']['NV fan 5_AC'][item_]
    
    text_height = bar_height + (y_range * 0.01)
    plt.text((3) * 0.5, text_height, str(round(p,4)), ha='center', va='bottom', c='k')

    plt.show()
```
