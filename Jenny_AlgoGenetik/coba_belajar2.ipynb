```python
def mutation(pop):
    popm = pop.copy()
    
    for i in range(10):
        position = random.sample(range(1,10),9)
        a = position[0]
        b = position[1]
        temp = popm.loc[i][a]
        popm.loc[i][a] = popm.loc[i][b]
        popm.loc[i][b] = temp
    return popm
```

```python
popm =mutation(pop)
popm
```

```python
def combinePopulation(pop, popm, x):
    popAll = pop.copy()
    popAll = popAll.append(x)
    popAll = popAll.append(popm)

    popAll = popAll.drop_duplicates()

    popAll.index = range(len(popAll))

    return popAll
```

```python
combinePopulation = combinePopulation(pop, popm, x)
combinePopulation
```
