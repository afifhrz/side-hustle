# PROBLEM 1

```python
fhand = open('worker.txt')
data = []
for line in fhand:
    line = line.split()
    data.append(line)
data
```

```python
from itertools import permutations
efficiency = []
permutate = list(permutations(range(8)))
for permute in permutate:
    temp = 0
    for i,number in enumerate(permute):
        temp += float(data[i][number])
    efficiency.append(temp)

print(max(efficiency),permutate[efficiency.index(max(efficiency))])
```

# PROBLEM 2

## using list

```python
fhand = open('exam.txt')
data = []
for line in fhand:
    line = line.split('\t')
    line[3] = line[3].replace('\n','')
    data.append(line)
data.pop(0)
data
```

```python
id = 1
append_count = 1
result = []
status_append = 1 
temp = []

for score in data:
    if status_append == 1:
        temp.append(score[0])
        temp.append(score[1])
        temp.append(score[3])
        status_append *= -1
    else:
        temp.append(score[3])
    append_count+=1
        
    if append_count == 4:
        append_count = 1
        id+=1
        result.append(temp)
        temp = []
        status_append *= -1

result
```

## using pandas

```python
import pandas as pd
data=pd.read_csv('exam.txt', sep='\t')
data = data.to_numpy()
data
```

```python
import numpy as np
id = 0
col = 2
result = np.zeros((14,5))
result = result.astype('O')
for score in data:
    result[id][0]=score[0]
    result[id][1]=score[1]
    result[id][col]=score[3]
    col+=1
    if col == 5:
        col = 2
        id+=1

pd.DataFrame(result, columns=['ID','gender','scores1','scores3','scores3'])
```

# PROBLEM 3

## using numpy

```python
import numpy as np
fhand = open('college.txt')
data = []
for line in fhand:
    line = line.split(' ')
    line[5] = line[5].replace('\n','')
    data.append(line)

data = np.array(data)
data
```

```python
result = []
for i in range(10):
    for student in data:
        if int(student[2])==i+1:
            result.append([i+1,int(student[0]),student[1]])

result
```

```python
import pandas as pd
data_2 = pd.DataFrame(result, columns=['College ID', 'Student Number', 'Score'])
for i in range(10):
    display(data_2[data_2['College ID']==i+1].sort_values('Score',ascending=False).head(10))
```

```python
import pandas as pd
data=pd.read_csv('college.txt', sep=' ', header=None)
data = data.to_numpy()
data
```
