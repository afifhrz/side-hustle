```python
import pandas as pd
data = pd.read_csv('lexicon.txt', header=None, sep=' ', names = ['type','len','word','pos','stemmed', 'prior_polarity'])
data
```

```python
data_used = data[['word','prior_polarity']]
data_used
```

```python
data_used_array = data_used.to_numpy()
data_used_array
```

```python
review = pd.read_csv('review.txt', header=None, names=['sentiment_text'])
review
```

```python
review_array = review.to_numpy()
review_array.shape
```

```python
sentiment_score=[]
for row_review_array in range(len(review_array)):
    sentiment_score.append(0)
    for row in range(len(data_used_array)):
        if data_used_array[row][0] in review_array[row_review_array][0].lower():
            if data_used_array[row][1]=='positive':
                sentiment_score[row_review_array]+=1
            elif data_used_array[row][1]=='negative':
                sentiment_score[row_review_array]-=1
            else:
                pass
sentiment_score
```

```python
review['sentiment_score'] = sentiment_score
review
```

```python
review['sentiment_score'].mean()
```
