```python
import pandas as pd
data=pd.read_csv('mpl.txt', sep='\t')
data
```

```python
new_data = data[['id','treatment']]
new_data
```

```python
array_to_loop = pd.array(data[['choice1','choice2','choice3','choice4','choice5'
,'choice6','choice7','choice8','choice9','choice10']], dtype='str')
array_to_loop
```

```python
switch_id = []
for row in range(array_to_loop.shape[0]):
    for col in range(array_to_loop.shape[1]):
        if array_to_loop[row][col]=='B':
            switch_id.append(col+1)
            break
switch_id
```

```python
new_data['switch_id']=switch_id
new_data.to_csv('switch.txt', sep = '\t', index = False)
```

```python
new_data.groupby('treatment').mean()
```

```python
new_data['condition'] = new_data['switch_id']>4
new_data.groupby('condition').count()
```
