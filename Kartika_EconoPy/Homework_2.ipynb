```python
import wooldridge as woo
wage = woo.data('wage2')
wage
```

```python
wage.describe()
```

```python
print("The average IQ in the sample is",wage.IQ.mean())
```

```python
print("The average salary in the sample is", wage.wage.mean())
```

```python
print("The sample standard deviation of IQ is", wage['IQ'].std())
```

```python
import statsmodels.formula.api as smf
reg = smf.ols('wage ~ IQ', data = wage)
results = reg.fit()
results.params
```

```python
print("The regression model is one-point increase in IQ changes wage by 116.99")
```

```python
parameter = results.params
IQ = 0
parameter[0]+parameter[1]*IQ
```
