# Number 3

```python
import wooldridge as woo
data_load = woo.data('bwght')
data_load
```

## 3A

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sturge
bins = np.round(1 + math.log(len(data_load),2)).astype(int)

plt.figure(figsize=(16,8))
plt.hist(data_load.bwght, alpha=.5, bins=bins)
plt.plot()
```

```python
# FD
def bins_fd(series):
    numerator = (len(series)**(1/3)*(series.max()-series.min()))
    denominator = (2*(data_load.bwght.quantile(.75)-data_load.bwght.quantile(.25)))

    return int(np.round(numerator/denominator))

plt.figure(figsize=(16,8))
plt.hist(data_load.bwght, alpha=.5, bins=bins_fd(data_load.bwght))
plt.plot()
```

## 3B

```python
import seaborn as sns

plt.figure(figsize=(10,6))
plt.title("Histogram")
sns.histplot(data_load.bwght, stat = "density")
sns.kdeplot(data_load.bwght, color = "red")
plt.plot()
```

## 3C

```python
import scipy.stats as stats

norm_shape = np.random.normal(0, 1, data_load.shape[0])

stats.probplot(norm_shape, dist = "norm", plot = plt)

plt.show()
```

## 3D

```python
women_smoke = data_load[(data_load['cigs']>0) & (data_load['male']==0)]['bwght']
women_not_smoke = data_load[(data_load['cigs']==0) & (data_load['male']==0)]['bwght']
data = [women_smoke, women_not_smoke]
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(data)
plt.ylabel("Cigs")
plt.xticks([1,2],["0","1"])
plt.show()
```

```python
import wooldridge as woo

data_load = woo.data('bwght')
data_load
```

## 4 A

```python
import statsmodels.formula.api as smf

reg = smf.ols('bwght ~ cigs', data = data_load)
result = reg.fit()

result.params
```

## 4 B

```python
parameter = result.params
```

```python
#cigs = 0
cigs = 0
parameter[0] + parameter[1]*cigs
```

```python
#cigs = 3
cigs = 3
parameter[0] + parameter[1]*cigs
```
