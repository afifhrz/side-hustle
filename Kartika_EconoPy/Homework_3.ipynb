```python
import wooldridge as woo
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# 1A

```python
data = woo.data("wage1")
data.describe()
```

```python
results = smf.ols('np.log(wage)~ educ + exper + tenure + female + married + numdep', data = data).fit()
results.summary()
```

```python
stats.jarque_bera(results.resid)
```

```python
# Reject Ho. The residuals from this regression does not normally distributed
```

# 1B

```python
xrange = np.linspace(data.exper.min(), data.exper.max(), 500)
xrange = pd.DataFrame(xrange, columns = ["exper"] )
```

```python
for key in ['educ', 'tenure', 'female', 'married', 'numdep']:
    xrange[key] = np.zeros(500)
xrange
```

```python
predictions = results.get_prediction(xrange)
# Generate table with intervals for each x

predictions = predictions.summary_frame(alpha=0.05)
```
